```python
import pandas as pd
import matplotlib.pyplot as plt

full_data_file=pd.read_csv('car_purchasing.csv',header=0,encoding='latin-1')
full_data_file
```

```python
#ExtrasiÃ³n de datos que utilizaremos
data_file=pd.read_csv('car_purchasing.csv',usecols=[5,8], encoding='latin-1')
data_file
```

```python
data_file.corr()
```

```python
mean_data = data_file.mean(0)
mean_data
```

```python
plt.scatter(data_file['annual Salary'],data_file['car purchase amount'],c='DarkBlue')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.legend()
plt.show()
mean_data = data_file.mean(0)
plt.scatter(mean_data['annual Salary'],mean_data['car purchase amount'],c='Red')
```

```python
from sklearn.linear_model import LinearRegression
# Crear un modelo de regresiÃ³n lineal
modelo = LinearRegression()

X = data_file['annual Salary'].values.reshape(-1, 1) # Variable independiente
Y = data_file['car purchase amount']#.values.reshape(-1, 1) # Variable dependependiente
# Ajustar el modelo a los datos
modelo.fit(X,Y)
```

```python
# Hacer predicciones
y_pred = modelo.predict(X)
```

```python
# Graficar los datos y la regresiÃ³n lineal
plt.scatter(X, Y, label='Datos',color='DarkBlue')
plt.plot(X, y_pred, color='red', label='RegresiÃ³n Lineal')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.legend()
plt.show()
```
